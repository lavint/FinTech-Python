```python
# Import libraries
from pathlib import Path
import csv

# Set file paths
menu_path = 'Resources/menu_data.csv'
sales_path = 'Resources/sales_data.csv'


# Initialize lists and dict
menu = []
sales = []
report = {}


# Read menu data into menu list
with open(menu_path, 'r') as menu_file:
    csvreader = csv.reader(menu_file, delimiter=',')
    
    # Skip the header row
    next(csvreader)
    
    # Iterate through each row
    for row in csvreader:
        #menu += [row]
        menu.append(row)

        
# Read sales data into sales list
with open(sales_path, 'r') as sales_file:
    csvreader = csv.reader(sales_file, delimiter=',')
    next(csvreader)
    #for row in csvreader:
        #sales.append(row)
    
    # Try another method to put csvreader into a list
    sales = list(csvreader)


# Loop through each line in sales    
for line in sales:
    
    # Initialize variables
    quantity = int(line[3])
    sales_item = line[4]
    
    # If sales item is not in report dict, initialize a key-value pair of matrices
    if sales_item not in report:
        report[sales_item] = {
            "01-count": 0,
            "02-revenue": 0,
            "03-cogs": 0,
            "04-profit": 0
        }
    
    # For each row in sales, loop through each line in menu to determine a match    
    for line in menu:
        
        # Initialize item variable from menu to capture ramen names
        item = line[0]
        
        # If item in menu equals sales_item in sales
        # Initialize price, cost, profit and track matrices for that item
        if item == sales_item:
            price = float(line[3])
            cost = float(line[4])
            profit = price - cost
            
            # Calculate cumulative sum for each matrix 
            report[sales_item]["01-count"] += quantity
            report[sales_item]["02-revenue"] += quantity * price
            report[sales_item]["03-cogs"] += quantity * cost
            report[sales_item]["04-profit"] += quantity * profit
            

# Output the result in a text file
with open('output.txt', 'w') as file:
    for key, value in report.items():      
        file.write(f'{key} {value}\n')
    
        
```
